```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
text= "square"
```

```python
wc = WordCloud(background_color="white",repeat=True,max_words=200)
```

```python
wc.generate(text)
```

```python
plt.axis("off")
plt.imshow(wc,interpolation="bilinear")
```

```python
import numpy as np
x,y = np.ogrid[:300,:300]
```

```python
mask = (x-150)**2 + (y-150)**2> 130**2
mask=255*mask.astype(int)
```

mask

```python
mask
```

```python
wc = WordCloud(background_color="white",repeat=True,max_words=200, mask=mask)
wc.generate(text)
plt.axis("off")
plt.imshow(wc,interpolation="bilinear")
```
