```python
from PIL import Image
import numpy as np
alice_mask = np.array(Image.open("C:\\Users\\wow\\Desktop\\alice_mask.png"))
```

```python
stopwords = set(STOPWORDS)
stopwords.add("said")
```

```python
wc = WordCloud(background_color="white",mask=alice_mask,max_words=10000,stopwords=stopwords,max_font_size=40,contour_width=3,contour_color="red")
wc.generate(alice)
plt.figure(figsize=(100,100))
plt.axis("off")
plt.imshow(wc,interpolation="bilinear")
```
