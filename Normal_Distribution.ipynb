```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#seed the randm number generater
np.random.seed(34589)
#주사위 50번던진후 눈평균
means = [np.mean(np.random.randint(1,7,50))for _ in range(10000)]
```

```python
#plot the distribution of sample mena
plt.hist(means)
```
