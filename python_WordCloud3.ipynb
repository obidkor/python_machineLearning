```python
text= kolaw.open('constitution.txt').read()
```

```python
from PIL import Image
import numpy as np
t=Okt()
token=t.nouns(text)
mask = np.array(Image.open("C:\\Users\\wow\\Desktop\\korea_mask.jpg"))
```

```python
stop_words=['제','월','일','조','수','때','그','이','바','및','안','정','위','등','항','것','관','인','둘','장']
token=[each_word for each_word in token if each_word not in stop_words]
```

```python
token
```

```python
ko = nltk.Text(token,name="대한민국헌법")
ko.vocab().most_common(100)
```

```python
data = ko.vocab().most_common(500)
tmp_data = dict(data)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
path='C:\\Windows\\SysWOW64\\Fonts\\NanumGothicExtraBold.ttf'
wc = WordCloud(background_color="white",relative_scaling=0,repeat=True,max_words=500,font_path=path,mask=mask,stopwords=stop_words)
plt.figure(figsize=(20,20))
wc.generate_from_frequencies(tmp_data)
plt.axis("off")
plt.imshow(wc,interpolation="bilinear")
```
